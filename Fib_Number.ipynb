Task 1：

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)

def fibonacci_dp(n, memo):
    if n <= 1:
        return n
    if n not in memo:
        memo[n] = fibonacci_dp(n - 1, memo) + fibonacci_dp(n - 2, memo)
    return memo[n]

import time
import matplotlib.pyplot as plt

recursive_times = []
dp_times = []

for n in range(10, 101, 10):
    if n > 50:
        recursive_times.append(float('inf'))
    else:
        start_time = time.time()
        fibonacci_recursive(n)
        end_time = time.time()
        recursive_times.append(end_time - start_time)

    start_time = time.time()
    fibonacci_dp(n, {})
    end_time = time.time()
    dp_times.append(end_time - start_time)

x = list(range(10, 101, 10))
plt.plot(x, recursive_times, label='Recursive')
plt.plot(x, dp_times, label='Dynamic Programming')
plt.xlabel('n')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Fibonacci Methods')
plt.legend()
plt.show()
```

```python
def find_max_recursive():
    n = 1
    while True:
        try:
            fibonacci_recursive(n)
            if n >= 50:
                recursive_times[n//10 - 1] = 1000
                break
            n += 1
        except (RecursionError, MemoryError):
            break
    return n - 1

max_n = find_max_recursive()
print(f"Maximum value of n for recursive method: {max_n}")

try:
    fibonacci_dp(max_n + 1, {})
    print(f"Dynamic programming method was able to compute F({max_n + 1}) without crashing.")
    print(f"And the {max_n + 1}th of Fibanocci Number is {fibonacci_dp(max_n+1, {})}")
except (RecursionError, MemoryError):
    print(f"Dynamic programming method also crashed when computing F({max_n + 1}).")
```
